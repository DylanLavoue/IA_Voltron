```python
from src.data.file_manager import FileManager
weatherManager = FileManager()
color_pal = sns.color_palette()
```

```python
raw_dir_path = weatherManager.data_dir / 'raw'

raw_path = str(raw_dir_path / 'theo_data.csv')
df_vigne = pd.read_csv(raw_path)
# df_vigne = df_vigne.reset_index()

#on rename index par Year
# df_vigne = df_vigne.rename(columns={'index': 'ds'})
df_vigne = df_vigne.rename(columns={'0': 'Weather'})
df_vigne
```

```python
raw_dir_path = weatherManager.data_dir / 'raw'

raw_path = str(raw_dir_path / 'raw_data.csv')
df_meteo = pd.read_csv(raw_path)
df_meteo
```

```python
# Convert the column in df_vigne to float type
df_vigne['Weather'] = df_vigne['Weather'].astype(float)

# Perform the merge
merged_df = df_vigne.merge(df_meteo, how='right', left_index=True, right_index=True)[['AverageTemperature', 'Weather' ]]
merged_df
```

```python
from sklearn.model_selection import train_test_split
X = merged_df.copy()
y = merged_df['Weather']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
#show difference between tree that doesn't have stopping criteria
clf.predict_proba(X_test)
```

```python
accuracy_score(y_test, predictions)
```

```python
confusion_matrix(y_test, predictions)
```

```python
precision_score(y_test, predictions, average='micro')
```
